```python
import pandas as pd

#Create 2 dataframes #1 BEFORE #2 AFTER
before_df = pd.read_csv('SLD_Case study_Events_Before.csv')
after_df = pd.read_csv('SLD_Case study_Events_After.csv')
```

```python
# To get unique value stats of the dataframe, including NaN data
uniqueBefore = before_df.nunique(dropna=False)
uniqueAfter = after_df.nunique(dropna=False)
print("BEFORE SHEET\n",uniqueBefore)
print("\n\nAFTER SHEET\n",uniqueAfter)
```

```python
# Fixing the gender column data in the Before Sheet
# Create a new dataframe to clean 'Gender' column
newBeforeDF = before_df[['Identifier (Email/name)', 'Gender']].copy()


# Check number of rows 
len(newBeforeDF.index)


#Dumping dataframe to CSV
#newBeforeDF.to_csv('Before-Gender.csv', index=False)

#uniqueGenderValues = newBeforeDF['Gender'].unique()
#print(uniqueGenderValues)

newBeforeDF['Gender'] = newBeforeDF['Gender'].fillna('Unspecified')


#uniqueGenderValues2 = newBeforeDF['Gender'].unique()
#print(uniqueGenderValues2)

#array = ['Female', 'Male', 'Unspecified']
#newBeforeDF.loc[newBeforeDF['Gender'].isin(array)]

# Create a new sheet for Before & After- Gender column
newBeforeDF.to_csv('Before-Gender.csv', index=False)


# Check number of rows 
len(newBeforeDF.index)

```

```python
# Fixing the gender column data in the After Sheet
#NOTE:: The column name is different in before and after sheet 'Gender' in before 'gender' in after
newAfterDF = after_df[['Identifier (Email/name)', 'gender']].copy()
len(newAfterDF.index)

uniqueGenderValues = newAfterDF['gender'].unique()
print("Original Data for 'gender' column :",uniqueGenderValues)

# Replace 'nan' values with 'Unspecified'
newAfterDF['gender'] = newAfterDF['gender'].fillna('Unspecified')
# Replace 'male' values with 'Male'
newAfterDF.loc[(newAfterDF.gender == 'male'),'gender']='Male'
# Replace 'female' values with 'Female'
newAfterDF.loc[(newAfterDF.gender == 'female'),'gender']='Female'
# Replace 'other' values with 'Unspecified'
newAfterDF.loc[(newAfterDF.gender == 'other'),'gender']='Unspecified'
# Replace '******' values with 'Unspecified'
newAfterDF.loc[(newAfterDF.gender == '******'),'gender']='Unspecified'


uniqueGenderValues = newAfterDF['gender'].unique()
print("Post update data:",uniqueGenderValues)

# Correct the gender column name to 'Gender' and align it to the Before sheet column name.
newAfterDF = newAfterDF.rename({'gender': 'Gender'}, axis='columns')

#Dump the dataframe to CSV file
newAfterDF.to_csv('After-Gender.csv', index=False)
len(newAfterDF.index)

```

```python
# Fixing Education column in the Before Sheet
import numpy as np
before_df = pd.read_csv('SLD_Case study_Events_Before.csv')
eduBeforeDF = before_df[['Identifier (Email/name)', 'Education']].copy()
#len(eduBeforeDF.index)

uniqueEducationValues = eduBeforeDF['Education'].unique()
#print("Original Data for 'Education' column :",uniqueEducationValues)

#print(eduBeforeDF['Education'].value_counts())

#eduBeforeDF[eduBeforeDF['Education'].str.contains("PhD")]='PhD'

# Replace null/empty values with 'NaN'
eduBeforeDF.replace('', np.nan, inplace=True) 

#Replace NaN values as 'Missing'
eduBeforeDF['Education'] = eduBeforeDF['Education'].fillna('Missing')

#print(eduBeforeDF['Education'].value_counts())
eduBeforeDF['Education_v1']=pd.np.where(eduBeforeDF.Education.str.contains("phd|doctorate", case=False), "Doctorate",
                            pd.np.where(eduBeforeDF.Education.str.contains("master|masters|MBA|MS|MA|postgraduate", regex=True,case=False), "Masters",
                            pd.np.where(eduBeforeDF.Education.str.contains("Engg|Engineering|BSc|BA|BCOM|B.A|Bachelors|Bachelor|B.Sc|BS|B.S|Degree|University|B.E|BE|B.Tech|grad|graduate|BTECH|NUS|NTU|RMIT|SMU|ACCA|SIM|SUSS|uni|HKU|ITS|ITB|Universitas|Monash|Deakin|La Trobe|Swinburne|UTS|UNSW|ANU", regex=True,case=False), "Bachelors",
                            pd.np.where(eduBeforeDF.Education.str.contains("undergraduate|diploma|college|TAFE|School|year|VCE|Poly", regex=True,case=False), "Undergraduate",
                            pd.np.where(eduBeforeDF.Education.str.contains("Missing",regex=True, case=False), "Unspecified","Others")))))


uniqueEducationValues = eduBeforeDF['Education_v1'].unique()
#print("Original Data for 'Education' column :",uniqueEducationValues)

print(eduBeforeDF['Education_v1'].value_counts())
uniqueAfter = eduBeforeDF.nunique(dropna=False)
print("\n\nAFTER SHEET\n",uniqueAfter)

#Drop column 'Education'
eduBeforeDF=eduBeforeDF.drop(columns=['Education'], axis=1)
print(eduBeforeDF.head(10)) 
#Rename column 'Education_v1' to 'Education'
eduBeforeDF=eduBeforeDF.rename({'Education_v1': 'Education'}, axis='columns')

#Dump the dataframe to CSV file
eduBeforeDF.to_csv('Before-Education.csv', index=False)
len(eduBeforeDF.index)
```

```python
# Getting Country details from the 'timezone' column in the After Sheet

import numpy as np
cityAfterDF = after_df[['Identifier (Email/name)', 'timezone']].copy()
print(len(cityAfterDF.index))
#print(cityAfterDF)

#print(cityAfterDF['timezone'].value_counts())

# The 'timezone' column contains data of format 'Continent/City', 
#so we will split the data to get city and then evaluate based on the city values.

# To create a new data frame with split value in column data, lets call this as 'continentCityDF' dataframe
continentCityDF = cityAfterDF["timezone"].str.split("/", n = 1, expand = True)
  
# The first split of the column is the continent information, 
# so we will create a new column called 'Continent' from the continentCityDF data frame
cityAfterDF["Continent"]= continentCityDF[0]
  
# The last split of the column is the City information, 
# so we will create a new column called 'City' from the continentCityDF data frame
cityAfterDF["City"]= continentCityDF[1]


# lets create 'Country' column from the 'City' column

cityAfterDF['Country']=pd.np.where(cityAfterDF.City.str.contains("Singapore", case=False), "Singapore",
                    pd.np.where(cityAfterDF.City.str.contains("Melbourne|Sydney", regex=True, case=False), "Australia",
                    pd.np.where(cityAfterDF.City.str.contains("Auckland",  case=False), "New Zealand",
                    pd.np.where(cityAfterDF.City.str.contains("Hong_Kong",  case=False), "Hong Kong (China)",
                    pd.np.where(cityAfterDF.City.str.contains("Jakarta", case=False), "Indonesia",
                    pd.np.where(cityAfterDF.City.str.contains("Kuala_Lumpur",  case=False), "Malaysia",
                    pd.np.where(cityAfterDF.City.str.contains("Los_Angeles",  case=False), "USA",
                    pd.np.where(cityAfterDF.City.str.contains("Manila", case=False), "Philippines",
                    pd.np.where(cityAfterDF.City.str.contains("Lagos", case=False), "Nigeria",
                    pd.np.where(cityAfterDF.City.str.contains("Phnom_Penh", case=False), "Cambodia",
                    pd.np.where(cityAfterDF.City.str.contains("Johannesburg", case=False), "South Africa","Unspecified")))))))))))

print(cityAfterDF['Country'].value_counts())
#print(cityAfterDF.head(10))

#Dump the dataframe to CSV file
cityAfterDF.to_csv('After-City_Country.csv', index=False)
print(len(cityAfterDF.index))
```
